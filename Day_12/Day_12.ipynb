## Problem statement
---

This last day is basically a packing problem, where you're given different present geometries, a list of areas and a number of each presents to be placed in those areas.
Solving the general case for this question is actually very difficult computationally as there are simply too many different arrangements of ways you can fit the presents into an area. 

Instead, this one is solvable in the way you would compute a lower bound i.e. assume each present takes up 9 spaces and see if the total number of presents multiplied by nine fits within the area. This is because the inputs were designed 'nicely' enough.

```python
with open('puzzle_input.txt', 'r') as f:
    lines = f.read().split('\n\n')

lines = [line.splitlines() for line in lines]

shapes = lines[:6]

for i in range(len(shapes)):
    shapes[i][0] = shapes[i][0][:1]

grids = lines[6:][0]
grids = [w.split(':') for w in grids]
grids = [(w[0].split('x'),w[1][1:].split(' ')) for w in grids]
```

```python
def grid_area(dim: list[str]) -> int:
    area = 1
    for c in dim:
        area *= int(c)

    #print(f"Dim is {dim}, area is {area}")
    return area

def max_shape_area(shape_list: list[str]) -> int:
    total = 0
    for c in shape_list:
        total += int(c) * 9

    #print(f"Shape_list is {shape_list}, max_shape area is {total}")
    return total

total = 0
for grid in grids:
    dim, shape_list = grid
    if max_shape_area(shape_list) <= grid_area(dim):
        total += 1
print(total)
    
    
```
